```python
#Dependencies
import pandas as pd
import numpy as np
import scipy.stats as sp
from Bio.KEGG import REST
```

```python
hsa_gene_dict = pd.read_pickle('hsa_gene_dict.p')
Metabolomics_data = pd.read_pickle('Metabolomics_data.p')
swath_lod = pd.read_pickle('SWATH_LoD.p')

```

```python
cpd_pw_text = REST.kegg_link("pathway","cpd").read()
```

```python
cpd_pw_text[0:100]
```

```python
cpd_pw_text = cpd_pw_text.split('\n')
```

```python
cpd_pw_text[0:10]
```

```python
pathway_cpd = []
for item in cpd_pw_text:
    if '\t' in item:
        item_split = item.split('\t')
        cpd_string = item_split[0]
        cpd_string = cpd_string[4:]
        
        map_string = item_split[1]
        map_string = map_string[5:]
        
        curr_list = [cpd_string,map_string]
        pathway_cpd.append(curr_list)
```

```python
pathway_cpd[0:10]
```

```python
map_list = [curr_list[1] for curr_list in pathway_cpd]
map_list_unique= list(set(map_list))
map_list[0:10]
```

```python
pathway_names = REST.kegg_list("pathway").read()
pathway_names = pathway_names.split('\n')
pathway_list = []

for entry in pathway_names:
    entry_split = entry.split('\t')
    if len(entry_split)==2:
        map_str = entry_split[0][5:]
        
        path_str = entry_split[1]
        curr_list = [map_str,path_str]
        pathway_list.append(curr_list)
        
pathway_list
```

```python
len(map_list_unique)
```

```python
PEA_output = []

for map_idx, curr_map in enumerate(map_list_unique):
    print(map_idx)
    
    for entry in pathway_list:
        if curr_map == entry[0]:
            path_name = entry[1]
            break


    is_sig_in_pathway=0
    is_sig_not_pathway=0
    not_sig_in_pathway=0
    not_sig_not_pathway=0

    for met_inx, met_data in enumerate(Metabolomics_data):
        S1 = met_data['SHAM']
        S2 = met_data['RYGB']

        ttest_result = sp.ttest_ind(S1,S2)
        ttest_result = ttest_result[1]

        KEGG_id = met_data['KEGG_id']

        in_pathway =0
        is_sig=0

        for i in range(len(pathway_cpd)):
            if pathway_cpd[i][0] == KEGG_id and pathway_cpd[i][1] == curr_map:
                in_pathway = 1
                break

        if ttest_result<0.05:
            is_sig = 1

        if in_pathway ==1 and is_sig ==1:
            is_sig_in_pathway+=1
        elif in_pathway==1 and is_sig ==0:
            not_sig_in_pathway+=1
        elif in_pathway==0 and is_sig ==1:
            is_sig_not_pathway+=1
        elif in_pathway==0 and is_sig==0:
            not_sig_not_pathway+=1

    contingency_table = [[is_sig_in_pathway,is_sig_not_pathway],[not_sig_in_pathway,not_sig_not_pathway]]
    oddsratio,pvalue = sp.fisher_exact(contingency_table)
    
    output_list = [curr_map, path_name, pvalue]
    PEA_output.append(output_list)
```

```python
PEA_output_df=pd.DataFrame(PEA_output, columns=["KEGG MAP", "Pathway Name", "pvalue"])
PEA_output_df = PEA_output_df.sort_values(by="pvalue",ascending=True).reset_index(drop=True)

PEA_output_df.to_csv('PEA_output.csv')
PEA_output_df
```

```python
hsa_pathway = REST.kegg_link("pathway","hsa").read()
hsa_pathway = hsa_pathway.split('\n')
hsa_pathway_list = []
for entry in hsa_pathway:
    entry_split = entry.split('\t')
    if len(entry_split)==2:
        hsa_id = entry_split[0]
        path_id = "map" + entry_split[1][8:]
        curr_list = [hsa_id,path_id]
        hsa_pathway_list.append(curr_list)
```

```python
map_list = [entry[1] for entry in hsa_pathway_list]
```

```python
map_list_unique = list(set(map_list))
```

```python
hsa_gene_dict['LPGP1']
```

```python
PEA_output_SWATH = []

for map_idx, curr_map in enumerate(map_list_unique):
    print(map_idx)
    
    for entry in pathway_list:
        if curr_map == entry[0]:
            path_name = entry[1]
            break


    is_sig_in_pathway=0
    is_sig_not_pathway=0
    not_sig_in_pathway=0
    not_sig_not_pathway=0

    for met_inx, prot_data in enumerate(swath_lod):
        S1 = prot_data['SHAM']
        S2 = prot_data['RYGB']
        prot = str(prot_data['protein_name']).upper()
        
        if prot in hsa_gene_dict:
            hsa_id = hsa_gene_dict[prot]['hsa_id']
            ttest_result = sp.ttest_ind(S1,S2)
            ttest_result = ttest_result[1]



            in_pathway =0
            is_sig=0

            for i in range(len(hsa_pathway_list)):
                if hsa_pathway_list[i][0] == hsa_id and hsa_pathway_list[i][1] == curr_map:
                    in_pathway = 1
                    break

            if ttest_result<0.05:
                is_sig = 1

            if in_pathway ==1 and is_sig ==1:
                is_sig_in_pathway+=1
            elif in_pathway==1 and is_sig ==0:
                not_sig_in_pathway+=1
            elif in_pathway==0 and is_sig ==1:
                is_sig_not_pathway+=1
            elif in_pathway==0 and is_sig==0:
                not_sig_not_pathway+=1

    contingency_table = [[is_sig_in_pathway,is_sig_not_pathway],[not_sig_in_pathway,not_sig_not_pathway]]
    oddsratio,pvalue = sp.fisher_exact(contingency_table)
    
    output_list = [curr_map, path_name, pvalue]
    PEA_output_SWATH.append(output_list)
```

```python
PEA_output_SWATH_df = pd.DataFrame(PEA_output_SWATH, columns = ['KEGG_Map', 'Pathway Name', 'pvalue'])
PEA_output_SWATH_df = PEA_output_SWATH_df.sort_values(by=['pvalue'])
PEA_output_SWATH_df.to_csv('PEA_SWATH.csv')
```
