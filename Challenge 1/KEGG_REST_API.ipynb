#### KEGG REST API
Using the KEGG REST API (http://rest.kegg.jp/list/hsa). 

I obtained a gene dictionary where the keys are all of the unique gene names.
The entries are dictionaries with three fields:  
1. List of synonym gene names 
2. Gene description 
3. Hsa identifier 

The Dictionary is converted to a DataFrame to view.

```python
from Bio.KEGG import REST
import pandas as pd
```

```python
human_genes = REST.kegg_list("hsa").read()
```

```python
human_genes_lines = human_genes.split('\n')
```

```python
full_dict = {}
for i in range(0,len(human_genes_lines)): 
    curr_dict = {}
    row = human_genes_lines[i]
    row_tabsplit = row.split('\t')
    hsa_id = row_tabsplit[0]
    if len(row_tabsplit)==2:
        second_half = row_tabsplit[1] 
        #added to prevent index error
    second_half_split = second_half.split(';')
    if len(second_half_split)==2:
        gene_description = second_half_split[1] 
        #added to prevent index error
    else:
        gene_description = second_half_split[0]
    gene_names = second_half_split[0]
    gene_name_vec = gene_names.split(',')
    
    gene_name_vec_new = []
    for gene in gene_name_vec:
        gene_name_vec_new.append(gene.strip())
    gene_name_vec = gene_name_vec_new    
    gene_name_vec = [gene.strip() for gene in gene_name_vec]
    curr_dict = {"gene_synonyms":gene_name_vec, 
                 "hsa_id":hsa_id, 
                "gene_description":gene_description}
    for gene in gene_name_vec:
        full_dict[gene]=curr_dict
        
   
```

```python
gene_df = pd.DataFrame(full_dict)
gene_df.to_csv("gene_df.csv")
gene_df.head()
```
