### Challenge 3

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
List_of_samples = []
for j in range(10000):
    s = np.random.uniform(0,1,100)
    #makes 100 random numbers between 0 and 1
    sample_bin = []
    #temp variable
    for i in range(100):
        if s[i]>0.4:
            #giving the odds of 60%
            sample_bin.append('R')
            #List_of_samples.append(apple)
        else:
            sample_bin.append('G')
            #List_of_samples.append(apple)
    List_of_samples.append(sample_bin)
```

```python
Red_count=[]
for i in range(10000):
    Red_count.append(List_of_samples[i].count('R'))
    
    
```

```python
plt.hist(Red_count)
plt.xlabel('Red Marbles per 100 sample size')
plt.ylabel('Number of repititions')
plt.title('Challenge 3 Histogram')
```

### Challenge 3 part 2

```python
Total = 0
for i in range(10000):
    Total = Total+Red_count[i]
Average=Total/10000   
```

```python
summation= 0
for i in range(10000):
    summation = summation + (Red_count[i]-Average)**2
stdev = (summation/10000)**0.5    
```

```python
stdev2=(Average*(1-0.6))**0.5
```

```python
mu, sigma = Average, stdev2
```

```python
s = np.random.normal(Average, stdev2, 10000)
```

```python
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.xlabel('Red Marbles per 100 sample size')
plt.ylabel('Number of repititions')
plt.title('Challenge 3 Part 2 Histogram')
```
