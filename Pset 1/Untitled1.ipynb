```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from pprint import pprint
import seaborn as sns
from Bio.KEGG import REST
```

```python
human_genes = REST.kegg_list("hsa").read()
```

```python
human_genes_lines = human_genes.split('\n')
```

```python
len(human_genes_lines)
```

```python
List_of_Genes = []

for i,row in enumerate(human_genes_lines):
    if len(row)>0:
        
    
        row_tab_split = row.split('\t')
        hsa_id = row_tab_split[0].strip()
    
        second_half = row_tab_split[1].strip()
        second_half_split = second_half.split(';')
    
        gene_names = second_half_split[0].strip()
        gene_names_vec = gene_names.split(',')
        gene_names_vec = [gene.strip() for gene in gene_names_vec]
        
        gene_description = 'No Description'
        
        if len(second_half_split)>1:
            gene_description = second_half_split[1].strip()
        curr_dict = {}
        curr_dict ={"gene_names_vec":gene_names_vec,
                   "hsa_id":hsa_id,
                   "gene_description":gene_description
                   }
        List_of_Genes.append(curr_dict)
```

```python
gene_names = []
for entry in List_of_Genes:
    gene_vec = entry['gene_names_vec']
    for gene in gene_vec:
        gene_names.append(gene)
```

```python
gene_names_unique = list(set(gene_names))
```

```python
#gene_name_synonyms = []
#gene_description = []
#hsa_identifier = []
#gene_name_fromunique = []
my_dict = {}
for gene in gene_names_unique:
    for entry in List_of_Genes:
        if gene in entry['gene_names_vec']:
            #gene_name_fromunique.append(gene)
            #gene_name_synonyms.append(entry['gene_names_vec'])
            #gene_description.append(entry['gene_description'])
            #hsa_identifier.append(entry['hsa_id'])
            my_dict.update({gene:{'gene_synonyms':entry['gene_names_vec'],
                                 'gene_description':entry['gene_description'],
                                 'hsa_id':entry['hsa_id']}})
            break
            
```

```python
my_df = pd.DataFrame(my_dict)
```
